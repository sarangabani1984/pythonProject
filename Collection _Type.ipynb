```python
products_prices = {
    "laptop": 1200,
    "Smartphone": 800,
    "headphones": 150,
    "Monitor": 300,
    "Keyboard": 75,
    "usb": 75
}

user = input("Enter the product:").lower()

if user in products_prices:
    print(f"The price of {user} is ${products_prices[user]}")
else:
    print("Product not found")
```
